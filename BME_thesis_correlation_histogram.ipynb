```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
correl_a_soma = np.load('transient_correl_coeffs/cor_list_a_soma.npy', allow_pickle = True)
correl_dendrite = np.load('transient_correl_coeffs/cor_list_dendritic.npy', allow_pickle = True)
correl_b_soma = np.load('transient_correl_coeffs/cor_list_b_soma.npy', allow_pickle = True)
```

```python
correlation_array = np.concatenate((correl_a_soma, correl_dendrite, correl_b_soma), axis = 0)
c_a_copy = correlation_array.copy()
```

```python
def false_splitter(f_array):
    false_transients = [i for i in f_array if i < 0.5]
    return false_transients
        
def true_splitter(t_array):
    true_transients = [j for j in t_array if j > 0.5] 
    return true_transients
```

```python
false_transients = false_splitter(correlation_array)
true_transients = true_splitter(c_a_copy)
```

```python
plt.hist(false_transients, alpha=0.5, label='false transients')
plt.hist(true_transients, alpha=0.5, label='true transients')
plt.legend(loc='upper right')
plt.show()
```

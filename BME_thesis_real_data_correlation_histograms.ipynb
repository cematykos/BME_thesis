```python
import numpy as np
import matplotlib.pyplot as plt
import os

import matplotlib.lines as mlines
```

```python
def loader(name = '_seq_dend_cor.npy', seq_num = '1', truefalse = 'true'):
    empty_list = []
    if name == '_seq_soma_truecor.npy':
        for i in range(1,8):
            name = f'{i}_seq_soma_truecor.npy'
            if os.path.isfile(name):
                empty_list.append(np.load(name, allow_pickle = True))
    else:
        for i in range(1,7):
            name = f'{seq_num}_seq_dend_{i}_{truefalse}cor.npy'
            if os.path.isfile(name):
                empty_list.append(np.load(name, allow_pickle = True))
    return empty_list
```

```python
at = loader('_seq_dend_cor.npy', '1', 'true')
af = loader('_seq_dend_cor.npy', '1', 'false')
bt = loader('_seq_dend_cor.npy', '2', 'true')
bf = loader('_seq_dend_cor.npy', '2', 'false')
ct = loader('_seq_dend_cor.npy', '3', 'true')
cf = loader('_seq_dend_cor.npy', '3', 'false')
dt = loader('_seq_dend_cor.npy', '4', 'true')
df = loader('_seq_dend_cor.npy', '4', 'false')
et = loader('_seq_dend_cor.npy', '5', 'true')
ef = loader('_seq_dend_cor.npy', '5', 'false')
ft = loader('_seq_dend_cor.npy', '6', 'true')
ff = loader('_seq_dend_cor.npy', '6', 'false')
gt = loader('_seq_dend_cor.npy', '7', 'true')
gf = loader('_seq_dend_cor.npy', '7', 'false')
```

```python
soma = loader('_seq_soma_truecor.npy', '', '')
```

```python
somT = list(soma[0]) + list(soma[1]) + list(soma[2]) + list(soma[3]) + list(soma[4]) + list(soma[5]) + list(soma[6])
somF = []

dend1T = list(at[0]) + list(bt[0]) + list(ct[0]) + list(dt[0]) + list(et[0]) + list(ft[0]) + list(gt[0])
dend1F = list(af[0]) + list(bf[0]) + list(cf[0]) + list(df[0]) + list(ef[0]) + list(ff[0]) + list(gf[0]) 

dend2T = list(at[1]) + list(bt[1]) + list(ct[1]) + list(dt[1]) + list(et[1]) + list(ft[1]) + list(gt[1])
dend2F = list(af[1]) + list(bf[1]) + list(cf[1]) + list(df[1]) + list(ef[1]) + list(ff[1]) + list(gf[1])  

dend3T = list(at[2]) + list(bt[2]) + list(ct[2]) + list(dt[2]) + list(et[2]) + list(ft[2]) + list(gt[2])
dend3F = list(af[2]) + list(bf[2]) + list(cf[2]) + list(df[2]) + list(ef[2]) + list(ff[2]) + list(gf[2])  

dend4T = list(at[3]) + list(bt[3]) + list(ct[3]) + list(dt[3]) + list(et[3]) + list(gt[3])
dend4F = list(af[3]) + list(bf[3]) + list(cf[3]) + list(df[3]) + list(ef[3]) + list(gf[3])  

dend5T = list(at[4]) + list(bt[4]) + list(ct[4]) + list(dt[4]) + list(et[4]) + list(gt[4])
dend5F = list(af[4]) + list(bf[4]) + list(cf[4]) + list(df[4]) + list(ef[4]) + list(gf[4])  

dend6T = list(at[5]) + list(bt[5]) + list(ct[5]) + list(dt[5]) + list(et[5]) + list(gt[5])
dend6F = list(af[5]) + list(bf[5]) + list(cf[5]) + list(df[5]) + list(ef[5]) + list(gf[5])   
```

## Data analysis - correlation histograms of synthetic spatial transient profiles and the source profiles of a specific neural compartment.

In this notebook I plot the correaltional histogram of true and false on transients of the synthetic dataset.

```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
```

```python
asoma_true = np.load('Asoma_true.npy', allow_pickle = True)

dendrite_true = np.load('dendrite_true.npy', allow_pickle = True)
dendrite_false = np.load('dendrite_false.npy', allow_pickle = True)

bsoma_true = np.load('Bsoma_true.npy', allow_pickle = True)
bsoma_false = np.load('Bsoma_false.npy', allow_pickle = True)
```

```python
local_dends = [dendrite_true[6], dendrite_true[13],
                dendrite_true[28], dendrite_true[35],
                dendrite_true[36], dendrite_true[40], 
                dendrite_true[45], dendrite_true[50], 
                dendrite_true[51], dendrite_true[55], 
                dendrite_true[56], dendrite_true[62],
                dendrite_true[63], dendrite_true[64],
                dendrite_true[65], dendrite_true[73],
                dendrite_true[74]]
```

Creating the plot:

```python
fig = plt.figure(figsize=(13, 4.5))

rows = 1
columns = 3

somF = []

#1
fig.add_subplot(rows, columns, 1)
plt.hist(somF, bins = np.arange(0,1,0.01), alpha=0.5)
plt.hist(asoma_true, bins = np.arange(0,1,0.01), alpha=0.5)
plt.xlabel('Correlation', fontsize = 13)
plt.ylabel('Number of \n corr. coeffs.', fontsize = 13)
plt.title('A - synthetic A Soma', fontsize = 13)
plt.xlim(0.7,1)

#2
fig.add_subplot(rows, columns, 2)
plt.hist(dendrite_false, bins = np.arange(0,1,0.01), alpha=0.5)
plt.hist(dendrite_true, bins = np.arange(0,1,0.01), alpha=0.5)
plt.hist(local_dends, bins = np.arange(0,1,0.01), alpha=0.5, color = 'r')
plt.xlabel('Correlation', fontsize = 13)
plt.ylabel('Number of \n corr. coeffs.', fontsize = 13)
plt.title('B - synthetic A dendrite', fontsize = 13)
plt.xlim(0.4,1)

#3
fig.add_subplot(rows, columns, 3)
plt.hist(bsoma_false, bins = np.arange(0,1,0.01), alpha=0.5)
plt.hist(bsoma_true, bins = np.arange(0,1,0.01), alpha=0.5)
plt.xlabel('Correlation', fontsize = 13)
plt.ylabel('Number of \n corr. coeffs.', fontsize = 13)
plt.title('C - synthetic B soma', fontsize = 13)

one = mlines.Line2D([], [], color='orange', marker='s', ls='', alpha = 0.5)
two = mlines.Line2D([], [], color='cornflowerblue', marker='s', ls='', alpha = 0.5)
three = mlines.Line2D([], [], color='red', marker='s', ls='')


fig.legend(handles=[one, two, three], labels=['false transients', 'true transients', 'local transients'], loc = 'upper right', fontsize=8)

fig.tight_layout(pad=5.0)

plt.suptitle('Correlation of the source profiles \n and the transient profiles - synthetic data \n', fontsize=17, fontweight = 'bold')
```
